```python
import numpy as np
dfa1 = dfa[["Background","Neoplastic","Inflammatory","Connective","Dead","Non-Neoplastic Epithelial"]]
dfb1 = dfb[["Background","Neoplastic","Inflammatory","Connective","Dead","Non-Neoplastic Epithelial"]]

Backgrounda = list(dfa1["Background"])
BackgroundNa = np.zeros(len(Backgrounda))
for i in range(len(Backgrounda)):
    BackgroundNa[i] = Backgrounda[i]/np.sum(Backgrounda)
  
Neoplastica = list(dfa1["Neoplastic"])
NeoplasticNa = np.zeros(len(Neoplastica))
for i in range(len(Neoplastica)):
    NeoplasticNa[i] = Neoplastica[i]/np.sum(Neoplastica)
  
Inflammatorya = list(dfa1["Inflammatory"])
InflammatoryNa = np.zeros(len(Inflammatorya))
for i in range(len(Inflammatorya)):
    InflammatoryNa[i] = Inflammatorya[i]/np.sum(Inflammatorya)
 

Connectivea = list(dfa1["Connective"])
ConnectiveNa = np.zeros(len(Connectivea))
for i in range(len(Connectivea)):
    ConnectiveNa[i] = Connectivea[i]/np.sum(Connectivea)
 
Deada = list(dfa1["Dead"])
DeadNa = np.zeros(len(Deada))
for i in range(len(Deada)):
    DeadNa[i] = Deada[i]/np.sum(Deada)
 

Non_Neoa = list(dfa1["Non-Neoplastic Epithelial"])
Non_NeoNa = np.zeros(len(Non_Neoa))
for i in range(len(Non_Neoa)):
    Non_NeoNa[i] = Non_Neoa[i]/np.sum(Non_Neoa)
 
##################################################
Background = list(dfb1["Background"])
BackgroundN = np.zeros(len(Background))
for i in range(len(Background)):
    BackgroundN[i] = Background[i]/np.sum(Background)
  

Neoplastic = list(dfb1["Neoplastic"])
NeoplasticN = np.zeros(len(Neoplastic))
for i in range(len(Neoplastic)):
    NeoplasticN[i] = Neoplastic[i]/np.sum(Neoplastic)
 

Inflammatory = list(dfb1["Inflammatory"])
InflammatoryN = np.zeros(len(Inflammatory))
for i in range(len(Inflammatory)):
    InflammatoryN[i] = Inflammatory[i]/np.sum(Inflammatory)

    
Connective = list(dfb1["Connective"])
ConnectiveN = np.zeros(len(Connective))
for i in range(len(Connective)):
    ConnectiveN[i] = Connective[i]/np.sum(Connective)
 

Dead = list(dfb1["Dead"])
DeadN = np.zeros(len(Dead))
for i in range(len(Dead)):
    DeadN[i] = Deada[i]/np.sum(Dead)
 

Non_Neo = list(dfb1["Non-Neoplastic Epithelial"])
Non_NeoN = np.zeros(len(Non_Neo))
for i in range(len(Non_Neo)):
    Non_NeoN[i] = Non_Neo[i]/np.sum(Non_Neo)

    
data_a = [BackgroundNa,NeoplasticNa,InflammatoryNa,ConnectiveNa,DeadNa,Non_NeoNa]
data_b = [BackgroundN,NeoplasticN,InflammatoryN,ConnectiveN,DeadN,Non_NeoN]
```
